### Week 1: Python Basics and Data Structures

#### Topics Covered:

1. **Python Fundamentals:**
   - **Syntax and Variables:**
     - **Definition:** The basic rules of writing Python code and using variables to store data.
     - **Example:**
       ```python
       # Variable assignment
       x = 10
       y = "Hello, Python!"
       print(x)  # Output: 10
       print(y)  # Output: Hello, Python!
       ```

   - **Data Types:**
     - **Definition:** Different kinds of data that Python can handle, such as integers, floats, strings, and booleans.
     - **Example:**
       ```python
       # Data types
       x = 5         # Integer
       y = 3.14      # Float
       z = "Python"  # String
       a = True      # Boolean
       ```

2. **Control Flow:**
   - **If Statements:**
     - **Definition:** Conditional statements that allow the execution of different code blocks based on certain conditions.
     - **Example:**
       ```python
       # If statements
       age = 18
       if age >= 18:
           print("You are an adult.")
       else:
           print("You are a minor.")
       ```

   - **Loops:**
     - **Definition:** Constructs that allow repeated execution of a block of code.
     - **For Loop Example:**
       ```python
       # For loop
       for i in range(5):
           print(i)
       ```
     - **While Loop Example:**
       ```python
       # While loop
       count = 0
       while count < 5:
           print(count)
           count += 1
       ```

3. **Functions and Modules:**
   - **Functions:**
     - **Definition:** Reusable blocks of code that perform a specific task.
     - **Example:**
       ```python
       # Function definition
       def greet(name):
           return f"Hello, {name}!"

       # Function call
       print(greet("Alice"))  # Output: Hello, Alice!
       ```

   - **Modules:**
     - **Definition:** Files containing Python code that can be imported and used in other Python scripts.
     - **Example:**
       ```python
       # math module example
       import math

       print(math.sqrt(16))  # Output: 4.0
       ```

4. **Data Structures:**
   - **Lists:**
     - **Definition:** Ordered, mutable collections of items.
     - **Example:**
       ```python
       # List example
       fruits = ["apple", "banana", "cherry"]
       fruits.append("date")
       print(fruits)  # Output: ['apple', 'banana', 'cherry', 'date']
       ```

   - **Tuples:**
     - **Definition:** Ordered, immutable collections of items.
     - **Example:**
       ```python
       # Tuple example
       dimensions = (1920, 1080)
       print(dimensions)  # Output: (1920, 1080)
       ```

   - **Sets:**
     - **Definition:** Unordered collections of unique items.
     - **Example:**
       ```python
       # Set example
       unique_numbers = {1, 2, 3, 4, 4, 5}
       print(unique_numbers)  # Output: {1, 2, 3, 4, 5}
       ```

   - **Dictionaries:**
     - **Definition:** Collections of key-value pairs.
     - **Example:**
       ```python
       # Dictionary example
       student = {"name": "John", "age": 20, "major": "Computer Science"}
       print(student["name"])  # Output: John
       ```

5. **Introduction to List Comprehensions:**
   - **Definition:** A concise way to create lists.
   - **Example:**
     ```python
     # List comprehension example
     squares = [x ** 2 for x in range(10)]
     print(squares)  # Output: [0, 1, 4, 9, 16, 25, 36, 49, 64, 81]
     ```
